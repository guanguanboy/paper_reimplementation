scikit-image是一个图像处理算法的集合。它是基于scipy的一款图像处理包，它将图片作为numpy数组进行处理，方便进行后续运算。



```python
import skimage
from skimage import util
import numpy as np
import matplotlib.pyplot as plt

from skimage import util
from skimage import io

img = io.imread('000001.jpg')
noise_gs_img = util.random_noise(img,mode='gaussian')

plt.imshow(noise_gs_img)

io.imsave('noise_000001.jpg', noise_gs_img)
```
